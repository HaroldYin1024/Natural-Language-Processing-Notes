# Conditional frequency distributions

# 1. Text corpora

The Brown corpus (1961) contians text from 500 sources, categorized by genre(e.g. news, editorial, hobbies, fiction, etc.

```python
import nltk
from nltk.corpus import brown
nltk.download('brown')
```

```python
brown.categories()
```

```python
brown.words(categories='news')
```

# 2. Conditional frequency distributions
Frequency distributions conditioned on categories

```python
cfd = nltk.ConditionalFreqDist((genre,word)
  for genre in brown.categories()
  for word in brown.words(categories=genre))
```

```python
cfd['news']
```

```python
cfd['science_fiction']
```

```python
from nltk.corpus import udhr
```

```python
languages=['English','German_Deutsch','Swahili_Kiswahili','Hungarian_Magyar']
```

```python
nltk.download('udhr')
```

## Can we find evidence of compounding?

```python
cfd=nltk.ConditionalFreqDist(
    (lang,len(word))
    for lang in languages
    for word in udhr.words(lang+'-Latin1')
)
```

```python
cfd.tabulate()
```

```python
cfd.plot()
```

# 3. Generating fake news
Using biagrams to produce random text

```python
s=['The','cat','sat','on','the','mat','.']
```

```python
list(nltk.bigrams(s))
```

```python
text=brown.words(categories='news')
bigrams=nltk.bigrams(text)
cfd=nltk.ConditionalFreqDist(bigrams)
```

```python
def generate_text(cfdist,word,num=50):
  for i in range(num):
    print(word,end=' ')
    word=cfdist[word].max()
```

```python
cfd.get('New')
```
